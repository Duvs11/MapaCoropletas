```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
gdf_localidades=gpd.read_file("/content/drive/MyDrive/POO/BarriosComunaSoacha.geojson")
gdf_localidades.plot()
```

```python
gdf_barrios=gpd.read_file("/content/drive/MyDrive/POO/Barrios_S.geojson")
gdf_barrios
```

```python
gdf_comuna_3 = gdf_barrios.loc[gdf_barrios['COM_MAPA'] == '7'].copy()
gdf_comuna_3.to_crs(epsg=9377,inplace=True)
gdf_comuna_3["Centroide"]=gdf_comuna_3.centroid
gdf_comuna_3.plot()
```

```python
mapa=gdf_comuna_3.plot(figsize=(10,15))
gdf_comuna_3["Centroide"].plot(ax=mapa, color="white")
```

```python
mapa2=gdf_comuna_3.plot(
    column="NOM_BARRIO",
    figsize=(30,15),
    edgecolor='black',
    cmap='gist_rainbow',
    legend=True,
    legend_kwds={
        'loc':'upper left'
    }
)
```
